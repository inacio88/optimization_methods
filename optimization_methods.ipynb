# Método da Bissecção

```python
#-----------------------Bissecção-----------------------------------------------
intervalo = b_final + a_inicio
c_metade = intervalo/2
y_ponto = (a_inicio + c_metade)/2
z_ponto = (b_final + c_metade)/2
```

```python
#---------------TESTE A FUNÇAO AQUI --------------------------------------------
teste_saida = expressao_symp.evalf(subs={x: 0.53})
print("Saida da função: {}".format(teste_saida))
```

```python
y_saida = expressao_symp.evalf(subs={x: y_ponto})
print('x1 é {}'.format(y_ponto))
print('f(x1) é {}'.format(y_saida))
z_saida = expressao_symp.evalf(subs={x: z_ponto})
print('x2 é {}'.format(z_ponto))
print('f(x2) é {}'.format(z_saida))
epsilon = z_ponto - y_ponto
print("Diferença: {}".format(epsilon))
```

```python
iteracao = 0
while epsilon > 0.01:
  iteracao = iteracao + 1
  print('--------------------------------------')
  print('Iteração: {}'.format(iteracao))
  #print("Diferença: {}".format(epsilon))
  print('x1 é {}'.format(y_ponto))
  print('f(x1) é {}'.format(y_saida))

  print('x2 é {}'.format(z_ponto))
  print('f(x2) é {}'.format(z_saida))

  if y_saida <= z_saida:
    b_final = z_ponto
    intervalo = b_final + a_inicio
    c_metade = intervalo/2
  else:
    a_inicio = y_ponto
    intervalo = b_final + a_inicio
    c_metade = intervalo/2


  epsilon = z_ponto - y_ponto

  
  y_ponto = (a_inicio + c_metade)/2
  z_ponto = (b_final + c_metade)/2

  

  y_saida = expressao_symp.evalf(subs={x: y_ponto})
  z_saida = expressao_symp.evalf(subs={x: z_ponto})
```
